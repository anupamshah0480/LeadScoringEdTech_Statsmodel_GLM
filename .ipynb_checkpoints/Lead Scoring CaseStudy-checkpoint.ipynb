```python
# creating a function for outlier removal

def remove_outlier(df,col):
    print(col)
    col_q1,col_q3,_col_IQR,lower_cutoff,upper_cutoff = 0,0,0,0,0
    data=pd.DataFrame()
    outlier_data=pd.DataFrame()
    data=df[col]
    

    col_q1 = data.quantile(q=.25)
    col_q3 = data.quantile(q=.75)
    col_IQR = col_q3 - col_q1
    lower_cutoff = col_q1 - (1.5 * col_IQR)
    upper_cutoff = col_q3 + (1.5 * col_IQR)
    
    mask1 = (df[col] > lower_cutoff) & (df[col] < upper_cutoff)

    if(lower_cutoff==0) | (upper_cutoff==0) | (lower_cutoff == upper_cutoff):
        print("Check cutoff values, it's 0 for: ",col)
        return df
        pass
    else:
        df = df[mask1]
        return df
```

```python
# Removing the outliers
for col in num_cols:
    leads = remove_outlier(leads,col)
```

### 7. Model Development

This is an iterative process,so there are a few points that we need to keep in mind:

    - There are 63 features, hence trying out different combinations will be tedious.
    - We will be using Sklearn RFE and statsmodels VIF in conjunction to filter out the significant features.
    - We will be using GLM with statsmodels for our logistic regression models.
    
Now before starting the development of model, lets devise a criteria to select a feature based on VIF and p-value:
    
    - Remove feature if high p-value
    - Features should have p-value<=0.05 [it signifies that features are statistically significant]
    - Features should have VIF <=5 [it signifies absense of multicollinearity]
    - ALWAYS REMOVE ONE FEATURE AT A TIME.
    

```python
# Lets first select 15 features filtered by RFE 

rfe = RFE(LogisticRegression(), 15)             
rfe = rfe.fit(X_train, y_train)
```

```python
rfe_df = pd.DataFrame({'Predictors': X_train.columns, 'Select': rfe.support_, 'Rank': rfe.ranking_})
rfe_df.sort_values(by='Rank')
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
# Lets build our first model: MODEL-I

X_train_sm = sm.add_constant(X_train[col])

log_reg_sm = sm.GLM(y_train,X_train_sm, family = sm.families.Binomial())

res = log_reg_sm.fit()

res.summary()
```

Here we can see there are two variables with higher p-value. Lets first remove 'Lead_Source_Referral Sites' as its has the highest p-value then we will go ahead with removing other one if required. 

```python
col = col.drop('Lead_Source_Referral Sites',1)
```

```python
# Lets again build the model: MODEL-II

X_train_sm = sm.add_constant(X_train[col])

log_reg_sm_II = sm.GLM(y_train,X_train_sm, family = sm.families.Binomial())

res = log_reg_sm_II.fit()

res.summary()
```

We still have a column with high p-value hence removing it.

```python
col = col.drop('Lead_Source_Organic Search',1)
```

```python
#MODEL-III

X_train_sm = sm.add_constant(X_train[col])

log_reg_sm_III = sm.GLM(y_train,X_train_sm, family = sm.families.Binomial())

res = log_reg_sm_III.fit()

res.summary()
```

Now that we have p-values of all the variable in the range, lets filter now based on VIF

```python
# creating a function to get VIF
def get_vif(df):
    vif_df = pd.DataFrame()
    vif_df['Features'] = df.columns
    vif_df['VIF'] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
    vif_df['VIF'] = round(vif_df['VIF'], 2)
    vif_df = vif_df.sort_values(by='VIF', ascending=False)
    vif_df = vif_df.reset_index(drop=True)
    return vif_df
```

```python
get_vif(X_train[col])
```

Now we have all the p-values and VIFs in control. Hence declaring the last model as our final model. Now we will be predicting the lead scores for Train data.

```python
y_train_pred = res.predict(X_train_sm)
y_train_pred[:10]
```

```python
y_train_pred = y_train_pred.values.reshape(-1)
y_train_pred[:10]
```

```python
y_train_final_df = pd.DataFrame({'Converted':y_train.values, 'Converted_prob':y_train_pred})
y_train_final_df['Prospect ID'] = y_train.index
y_train_final_df.head()
```

```python
y_train_final_df['Converted_pred'] = y_train_final_df.Converted_prob.map(lambda x: 1 if x > 0.5 else 0)
```

### Validating the model

```python
# Confusion matrix 
confusion = metrics.confusion_matrix(y_train_final_df.Converted, y_train_final_df.Converted_pred )
confusion
```

```python
# Let's check the overall accuracy.
print(metrics.accuracy_score(y_train_final_df.Converted, y_train_final_df.Converted_pred))
```

```python
# Creating a function for calculating different metrics

def scores(conf):
    TP = conf[1,1] # true positives
    TN = conf[0,0] # true negatives
    FP = conf[0,1] # false positives
    FN = conf[1,0] # false negatives
    
    sensitivity = TP/float(TP+FN)
    specificity = TN/float(TN+FP)
    precision = TP/float(TP+FP)
    accuracy = float(TP+TN)/float(TP+FP+TN+FN)
    
    print("Accuracy: ", round(accuracy,4))
    print("Sensitivity: ", round(sensitivity,4))
    print("Specificity: ", round(specificity,4))
    print("Precision: ", round(precision,4))
```

```python
scores(confusion)
```

```python
# Now lets plot the ROC Curve
# Creating a function to plot the ROC curve

def plot_roc( actual, probs ):
    fpr, tpr, thresholds = metrics.roc_curve( actual, probs,
                                              drop_intermediate = False )
    auc_score = metrics.roc_auc_score( actual, probs )
    plt.figure(figsize=(5, 5))
    plt.plot( fpr, tpr, label='ROC curve (area = %0.2f)' % auc_score )
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic example')
    plt.legend(loc="lower right")
    plt.show()
```

```python
fpr, tpr, thresholds = metrics.roc_curve( y_train_final_df.Converted, y_train_final_df.Converted_prob, drop_intermediate = False )
```

```python
plot_roc(y_train_final_df.Converted, y_train_final_df.Converted_prob)
```
